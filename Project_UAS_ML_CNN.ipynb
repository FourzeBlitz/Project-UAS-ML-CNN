# Normalization

Shuffle the data

```python
from sklearn.utils import shuffle

# Shuffle the training data
X_train, y_train = shuffle(X_train, y_train, random_state=42)
X_test, y_test = shuffle(X_test, y_test, random_state=42)
```

```python
X_train = X_train / 255
X_test = X_test / 255
```

# CNN
Pakai architecture AlexNet

```python
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.callbacks import LearningRateScheduler
import numpy as np
import math

# Set random seed for reproducibility
seed = 42
np.random.seed(seed)
tf.random.set_seed(seed)

# Initial learning rate
initial_lr = 0.01

# Define a learning rate schedule
def lr_schedule(epoch):
   return initial_lr * math.pow(0.1, math.floor(epoch / 10))

lr_scheduler = LearningRateScheduler(lr_schedule)

cnn = models.Sequential([
    # convo 1
   layers.Conv2D(filters=32, kernel_size=(5,5), activation='relu', input_shape=(28,28, 1)),
   layers.BatchNormalization(),
    # max pool 1
   layers.MaxPooling2D(pool_size=(2,2), strides=2),
    # convo 2
   layers.Conv2D(filters=64, kernel_size=(5,5), activation='relu', padding='same'),
   layers.BatchNormalization(),
    # max pool 2
   layers.MaxPooling2D(pool_size=(2,2), strides=2),
    # convo 3, 4, 5
   layers.Conv2D(filters=64, kernel_size=(5,5), activation='relu', padding='same'),
   layers.Conv2D(filters=64, kernel_size=(5,5), activation='relu', padding='same'),
   layers.Conv2D(filters=32, kernel_size=(5,5), activation='relu', padding='same'),
    # max pool 3
   layers.MaxPooling2D(pool_size=(2,2), strides=2),
    # dense with dropout
   layers.Flatten(),
   layers.Dense(128, activation='relu', kernel_initializer=tf.keras.initializers.GlorotUniform(seed=seed)),
   layers.BatchNormalization(),
   layers.Dropout(0.25, seed=seed),
   layers.Dense(64, activation='relu', kernel_initializer=tf.keras.initializers.GlorotUniform(seed=seed)),
   layers.BatchNormalization(),
   layers.Dropout(0.25, seed=seed),
   layers.Dense(10, activation='softmax', kernel_initializer=tf.keras.initializers.GlorotUniform(seed=seed))
])

# Use SGD optimizer with learning rate scheduler
sgd = optimizers.legacy.SGD(learning_rate=initial_lr, momentum=0.9)
cnn.compile(optimizer=sgd,
           loss='sparse_categorical_crossentropy',
           metrics=['accuracy'])

epochs = 15
```

```python
history = cnn.fit(X_train, y_train, epochs=epochs, validation_data=(X_test, y_test), callbacks=[lr_scheduler])
```

```python
from sklearn.metrics import confusion_matrix , classification_report
import numpy as np
y_pred = cnn.predict(X_test)
y_pred_classes = [np.argmax(element) for element in y_pred]

print("Classification Report: \n", classification_report(y_test, y_pred_classes))
```
